```python
#import libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the customer behaviour survey dataset.
df=pd.read_csv(r"C:\Users\hp\Downloads\Eezybuy Customer Behavior Survey.csv")
df
```

```python
#Clean the data.
#firstly check the info of the dataset.
df.info()
```

```python
#let's split the date from the time.

# Function to split date and time
def split_date_time(datetime_str):
    date_time_parts = datetime_str.split(' ')
    return date_time_parts[0]

# Apply the function to the Timestamp column to extract date
df['Date'] = df['Timestamp'].apply(split_date_time)

print(df)
```

```python
# Drop the Timestamp column
df.drop(columns=['Timestamp'], inplace=True)

print(df)
```

```python
df.head()
```

```python
# Drop rows with missing values
df.dropna(inplace=True)

df
```

```python
# Drop duplicate rows
df.drop_duplicates(inplace=True)

df
```

```python
# Function to categorize ages into groups
def categorize_age(age):
    if age <= 5:
        return '0-5'
    elif age <= 10:
        return '6-10'
    elif age <= 15:
        return '11-15'
    elif age <= 20:
        return '16-20'
    elif age <= 25:
        return '21-25'
    elif age <= 30:
        return '26-30'
    elif age <= 35:
        return '31-35'
    elif age <= 40:
        return '36-40'
    elif age <= 45:
        return '41-45'
    elif age <= 50:
        return '46-50'
    elif age <= 55:
        return '51-55'
    elif age <= 60:
        return '56-60'
    elif age <= 65:
        return '61-65'
    elif age <= 70:
        return '66-70'
    else:
        return '70+'

# Apply the function to create the 'age_group' column
df['age_group'] = df['age'].apply(categorize_age)
```

```python
df
```

```python
import seaborn as sns
```

```python
#Gender Analysis
#Plot purchase frequency distribution by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Purchase_Frequency', hue='Gender', data=df, palette='pastel')
plt.title('Purchase Frequency Distribution by Gender')
plt.xlabel('Purchase Frequency')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
#Visualize Age distribution

# Define age group ranges
age_group_labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61-65', '66-70']
age_group_ranges = [(0, 5), (6, 10), (11, 15), (16, 20), (21, 25), (26, 30), (31, 35), (36, 40), (41, 45), (46, 50), (51, 55), (56, 60), (61, 65), (66, 70)]

# Categorize ages into age groups
df['age_group'] = pd.cut(df['age'], bins=[x[0] for x in age_group_ranges] + [float('inf')], labels=age_group_labels, right=False)

# Plot histogram of age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=[x[0] for x in age_group_ranges] + [float('inf')], color='pink', edgecolor='black')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.xticks([x[0] for x in age_group_ranges], age_group_labels, rotation=45)
plt.show()
```

```python
# Group by age group and purchase frequency, and count the number of customers
grouped_df = df.groupby(['age_group', 'Purchase_Frequency']).size().unstack(fill_value=0)

# Plot the distribution
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', stacked=True)
plt.title('Distribution of Customers Across Age Groups and Purchase Frequencies')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.legend(title='Purchase Frequency')
plt.show()
```

```python
# Plot the distribution of shopping satisfaction levels

plt.figure(figsize=(8, 5))
df['Shopping_Satisfaction'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribution of Shopping Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```
